```python
# Solução do exercício 1
import Aulan6

Menu = str(""" Escolha uma das seguintes opções:
0 - Sair
1 - Carregar os dados para uma lista
2 - Número de obras no dataset
3 - Tabela com o título da obra, a sua descrição, o seu compositor e ano de criação
4 - Lista de tuplos (título,ano) ordenada alfabeticamente por título
5 - Lista de tuplos (título,ano) ordenada crescentemente por ano
6 - Lista ordenada dos compositores
7 - Distribuição das obras por período
8 - Distribuição das obras por ano
9 - Distribuição das obras por compositor 
10 - Gráficos com as distribuições
11 - Lista dos compositores em que cada compositor tem a ele associado uma lista dos títulos das obras que compôs """)

print(Menu)

n = 11

while n!=0:
    obras = Aulan6.lerObras()
    n = int(input("Introduza a opção que escolheu:"))
    print ("Opção escolhida:", n)
    if n == 1:
        print(f"{'Lista de dados':=^125}\n")
        print(Aulan6.lerObras())
        print()
    if n == 2:
        print("Número de obras no dataset:", Aulan6.contarObras(obras))
        print()
    if n == 3:
        print(f"{'Tabela organizada':=^83}")
        print(Aulan6.imprimeObras(obras))
        print()
    if n == 4:
        print(f"{'Lista de tuplos (Título,ano) ordenada alfabeticamente por título':=^125}")
        print(Aulan6.ordenar(obras))
        print()
    if n == 5:
        print(f"{'Lista de tuplos (Título,ano) ordenada crescentemente por ano':=^125}")
        print(Aulan6.ordenar1(obras))
        print()
    if n == 6:
        print(f"{'Lista ordenada dos compositores' :=^125}")
        print(Aulan6.ordem2(obras))
    if n == 7:
        print(f"{'Distribuição por período':=^125}")
        print(Aulan6.distNP(obras))
        print()
    if n == 8:
        print(f"{'Distribuição por ano':=^125}")
        print(Aulan6.distNT(obras))
        print()
    if n == 9:
        print(f"{'Distribuição por compositores':=^125}")
        print(Aulan6.distNC(obras))
        print()
    if n == 10:
        print("Possíveis distribuições: 1 - Período, 2 - Ano, 3 - Compositor")
        m = 4
        while m != 0:
            m = int(input("Introduza o número da distribuição que pretende observar:"))
            if m == 1:
                print("Distribuição por período")
                ob1 = Aulan6.distPeriodo(obras)
                print(Aulan6.distribuico(ob1))
                print()
            if m == 2:
                print("Distribuição por ano")
                ob2 = Aulan6.distNT(obras)
                print(Aulan6.distribuico1(ob2))
                print()
            if m == 3:
                print("Distribuição por compositor")
                ob3 = Aulan6.distNC(obras)
                print(Aulan6.distribuico2(ob3))
                print()
        if m == 0:
            print("A aplicação fechou")
    if n == 11:
        print(f"{'Lista dos compositores em que cada compositor tem a ele associado uma lista dos títulos das obras que compôs':=^125}")
        ob4 = Aulan6.distNC(obras)
        print(Aulan6.COMPOBRAS(ob4))
    
if n == 0:
    print("Saiu da aplicação")
```
